```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import scipy as sc
```

```python
fig, ax = plt.subplots(1, 1)
dfn, dfd = 29,18
r = np.random.f(dfnum=dfn, dfden=dfd, size=100000)
sns.histplot(r)
```

```python
mean, var, skew, kurt = sc.stats.f.stats(dfn,dfd, moments='mvsk')
x = np.linspace(sc.stats.f.ppf(0.05, dfn, dfd),
                sc.stats.f.ppf(0.95, dfn, dfd), 100)
ax.plot(x, sc.stats.f.pdf(x, dfn, dfd),
       'r-', lw=5, alpha=0.6, label='f pdf')
```

```python
rv = sc.stats.f(dfn, dfd)
ax.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')
vals = sc.stats.f.ppf([0.001, 0.5, 0.999], dfn, dfd)
np.allclose([0.001, 0.5, 0.999], sc.stats.f.cdf(vals, dfn, dfd))
```
