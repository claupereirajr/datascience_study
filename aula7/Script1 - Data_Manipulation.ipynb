```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
dataset_raw = pd.read_json("https://my.api.mockaroo.com/employ_quiz.json?key=9879a790")
dataset_raw.head()
```

```python
dataset_init = dataset_raw.copy()
dataset_init.head()
```

```python
columns_drop = ['firstName', 'email']
columns_drop
```

```python
dataset_init = dataset_init.drop(columns=columns_drop)
dataset_init.head()
```

```python
id_vars = list(dataset_init.columns)[:5]
value_vars = list(dataset_init.columns)[5:]
print("id_vars", id_vars)
print("value_vars", value_vars)
```

```python
dataset_melt = dataset_init.melt(id_vars=id_vars, value_vars=value_vars, var_name="Question + Sub" , value_name="Answer")
print("dataset_melt: ", len(dataset_melt))
dataset_melt.head()
```

```python
respondents = dataset_melt[dataset_melt["Answer"].notna()]
respondents = respondents.groupby("Question + Sub")["userId"].nunique().reset_index()
respondents.rename(columns={"userId":"Total Answer"}, inplace=True)
respondents
```

```python
dataset_merged = pd.merge(left=dataset_melt, right=respondents, how="left", left_on="Question + Sub", right_on="Question + Sub")
print("dataset_melt",len(dataset_melt))
print("dataset_merged",len(dataset_merged))
dataset_merged.head()
```

```python
same_answer = dataset_melt
same_answer = same_answer.groupby(["Question + Sub", "Answer"])["userId"].nunique().reset_index()
same_answer.rename(columns={"userId":"Same Anwer"}, inplace=True)
same_answer
```

```python
dataset_merged_two = pd.merge(left=dataset_merged, right=same_answer, how="left", left_on=["Question + Sub", "Answer"], right_on=["Question + Sub", "Answer"])
print("dataset_melt", len(dataset_melt))
print("dataset_merged",len(dataset_merged_two))
dataset_merged_two
```
